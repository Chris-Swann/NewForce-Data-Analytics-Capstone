```python
import pandas as pd
import glob
import os

# Set the folder path
folder_path = '../data/raw/utmb_format'

# Find all CSV files in the folder
csv_files_utmb_format = glob.glob(os.path.join(folder_path, '*.csv'))

# Read and combine all CSVs
df_list = []
for file in csv_files_utmb_format:
    temp_df = pd.read_csv(file)
    df_list.append(temp_df)

# Combine into one DataFrame
combined_df = pd.concat(df_list, ignore_index=True)
```

```python
combined_df.head(28)
```

```python
# Step 1: Filter out DNFs only
df_finished = combined_df[combined_df['Status'].str.upper() != 'DNF'].copy()

# Step 2: Sort by Race_ID, Gender, and Time
df_finished = df_finished.sort_values(by=['Race_ID', 'Gender', 'Time'])

# Step 3: Assign Gender_Rank
df_finished['Gender_Rank'] = df_finished.groupby(['Race_ID', 'Gender']).cumcount() + 1

# Step 4: Merge back using Race_ID and Name
df = combined_df.merge(
    df_finished[['Race_ID', 'Name', 'Gender', 'Time', 'Gender_Rank']],
    on=['Race_ID', 'Name', 'Gender', 'Time'],
    how='left'
)

```

```python
def map_utmb_age_category(age):
    if age < 18:
        return 'U18'
    elif age < 20:
        return 'U20'
    elif age <= 34:
        return '20-34'
    elif age <= 39:
        return '35-39'
    elif age <= 44:
        return '40-44'
    elif age <= 49:
        return '45-49'
    elif age <= 54:
        return '50-54'
    elif age <= 59:
        return '55-59'
    elif age <= 64:
        return '60-64'
    elif age <= 69:
        return '65-69'
    elif age <= 74:
        return '70-74'
    elif age <= 79:
        return '75-79'
    else:
        return '80'
```
